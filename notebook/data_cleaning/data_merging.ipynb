## **Merging the cleaned flats and houses Dataset**

```python
# importing the libraries
import pandas as pd
```

```python
# loading the datasets
flats = pd.read_csv('../data/flats_cleaned.csv')
houses = pd.read_csv('../data/house_cleaned.csv')
```

```python
# concatinating the datasets
data = pd.concat([flats, houses], ignore_index=True)
```

```python
data = data.sample(data.shape[0],ignore_index=True)
```

```python
# exporting the merged dataset
data.to_csv('../data/gurgaon_properties.csv', index=False)
```
