## **Preprocessing**

#### **1. `society` Column**

```python
data['society'].value_counts()
```

There are many values that ends with a `★`, which is not needed.

```python
data['society'] = data['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

#### **2. `price` Column**

```python
data['price'].value_counts()
```

In `price` column some flats have there price in Lacs and some have there price in Crore. Thus converting the flats price in a single unit that is Crore. 

Removing flats that have values as `Price on Request`. 

```python
data = data[data['price']!='Price on Request']
```

```python
# Creating a price convertor function
def price_converter(x):
    """
    This function converts price amounts into Crores.
    """
    # If the input is already a float, return it directly
    if isinstance(x, float):
        return x

    # Split the string into parts
    parts = x.split(' ')

    # Check if the split resulted in at least two parts
    if len(parts) < 2:
        return None  # Or handle it in another way, like returning 0

    # Extract the numeric value and unit
    numeric_value = float(parts[0])
    unit = parts[1]

    # Convert based on the unit
    if unit == 'Lac':
        return round(numeric_value / 100, 2)
    elif unit == 'Crore':
        return round(numeric_value, 2)
    else:
        return None  # Handle unexpected units
```

```python
# Applying the price converter function to the price feature
data['price'] = data['price'].apply(price_converter)
```

```python
data.head()
```
